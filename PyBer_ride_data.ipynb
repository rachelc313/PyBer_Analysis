```python
#Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import statistics
```

```python
#Create the path and import the data
pyber_ride_data = os.path.join('Resources/PyBer_ride_data.csv')
pyber_ride_df = pd.read_csv(pyber_ride_data)
```

```python
# Plot data on a line graph
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.

# Need an array that is the length of the DataFrame. Use arange function for the length of the pyber_ride_df
x_axis = np.arange(len(pyber_ride_df))

#Use list comprehension to set the tick locations for every value, or month in the x-axis.
tick_locations = [value for value in x_axis]

#Plot the data
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")

#Add the xticks() function to pass in the tick_locations and the labels of the x-ticks, pyber_ride_df["Month"]
plt.xticks(tick_locations, pyber_ride_df["Month"])

#Show graph
plt.show
```

```python
# Create plot
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", color="skyblue")

# Get the standard deviation of the values in the y-axis
#Define x_axis
x_axis = pyber_ride_df.loc[:, "Month"]

#Define y_axis
y_axis = pyber_ride_df.loc[:, "Avg. Fare ($USD)"]

#Define stdev
stdev = statistics.stdev(y_axis)

# Add error bars with caps
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3, color="black")

#Rotate the labels on the x-axis to horizontal
plt.xticks(rotation=0)

# Set the  y-axis to increment every $5
plt.yticks(np.arange(0, 51, step=5.0))

plt.show
```
